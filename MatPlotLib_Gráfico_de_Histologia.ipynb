```python
# Importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando uma variavel X com numeros aleatorios
x = np.random.rand(2000)
```

```python
# Criando a histologia
plt.hist(x)
plt.xlabel('x')
plt.ylabel('contagem')
plt.show()
```

```python
# Alterando o numero de Bins
plt.hist(x,bins=10)
plt.xlabel('x')
plt.ylabel('Contagem')
plt.title('bins=10')
plt.show()
```

```python
plt.hist(x,bins=30)
plt.xlabel('x')
plt.ylabel('Contagem')
plt.title('bins=30')
plt.show()
```

```python
plt.hist(x,bins=60)
plt.xlabel(x)
plt.ylabel('contagem')
plt.title('bins=60')
plt.show()
```

```python
#  Alterando a cor
plt.hist(x,bins=60,color='green')
plt.xlabel('x')
plt.ylabel('Contagem')
plt.title('bins=60')
plt.show()

```

```python
plt.hist(x,bins=60,color='green', edgecolor='blue')
plt.xlabel('x')
plt.ylabel('Contagem')
plt.title('bins=60')
plt.show()
```

```python
# Mutiplos histogramas na mesma imagem
x1 = np.random.normal(0,1,size = 800)
x2 = np.random.normal(3,1,size = 800)
x3 = np.random.normal(5,1,size = 800)
```

```python
plt.hist(x1,bins=10,color='black')
plt.hist(x2,bins=10,color='blue')
plt.hist(x3,bins=10,color='green')
plt.xlabel(x)
plt.ylabel('contagem')
plt.title('Multiplos histogramas')
plt.show()
```

```python
# triplos histogramas na mesma imagem
x1 = np.random.normal(0,1,size = 800)
x2 = np.random.normal(3,1,size = 800)
x3 = np.random.normal(5,1,size = 800)
x4 = np.random.normal(7,1,size = 800)
```

```python
plt.hist(x1,bins=10,color='purple')
plt.hist(x2,bins=10,color='blue')
plt.hist(x3,bins=10,color='green')
plt.hist(x4,bins=10,color='orange')
plt.xlabel(x)
plt.ylabel('contagem')
plt.title('triplos histogramas')
plt.show()
```

```python
# Distribui√ßao cumulativa
```

```python
x = np.random.rand(2000)
```

```python
plt.hist(x,bins=20,color='violet', edgecolor='red',cumulative=True)
plt.xlabel('x')
plt.ylabel('Contagem')
plt.title('Histograma cumulativo')
plt.show()
```
