```python
 #Importando Bibliotecas

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando variavel x

x = np.linspace(0,5,11)
```

```python
# Criando funções y(x) e z(
 
y = np.power(x-2,2.0)
```

```python
z = np.exp(x/2.0)
```

```python
plt.subplot(2,1,1)
plt.plot(x,y)
plt.subplot(2,1,2)
plt.plot(x,z)
plt.show()
```

```python
# Nomeando eixo de cada gráfico

plt.subplot(2,1,1)
plt.plot(x,y)
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.subplot(2,1,2)
plt.plot(x,z)
plt.xlabel('Eixo x')
plt.ylabel('Eixo z')
plt.tight_layout()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.subplot(1,2,2)
plt.plot(x,z)
plt.xlabel('Eixo x')
plt.ylabel('Eixo z')
plt.tight_layout()
plt.show()
```

```python
# Usando notações de objeto

fig,ax = plt.subplots(nrows=1,ncols=2)
ax[0].plot(x,y)
ax[0].set_xlabel('Eixo x')
ax[0].set_ylabel('Eixo y')
ax[1].plot(x,z)
ax[1].set_xlabel('Eixo x')
ax[1].set_ylabel('Eixo z')
plt.tight_layout()
```
