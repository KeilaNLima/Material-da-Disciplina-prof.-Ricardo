```python
# Importando Bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando variavel x
x = np.linspace(0,5,11)
```

```python
# Criando funçoes y(x) e z(x)
y = np.power(x-2,2.0)
z = np.exp(x/2.0)

```

```python
# Criando gráfico em funçao de x
plt.plot(x,y)
plt.show()
```

```python
# Inserindo nomes nos eixos
plt.plot(x,y)
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.show()
```

```python
# Inserindo titulo no gráfico
plt.plot(x,y)
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('y(x)')
plt.show()
```

```python
# Dois gráficos na mesma imagem
plt.plot(x,z)
plt.plot(x,y)
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('y(x), z(x)')
plt.show()
```

```python
# Alterando cores
plt.plot(x,y,color='black')
plt.plot(x,z,color='red')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.show()
```

```python

# Inserindo legenda
plt.plot(x,y,color='green',label='y(x)')
plt.plot(x,z,color='red',label='z(x)')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.legend()
plt.show()
```
