```python
# Importando Bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando variavel X
x = np.linspace(0,5,11)
```

```python
# Criando fun√ßoes y(x) e z(x)
y  = np.power(x-2,2.0)
```

```python
z = np.exp(x/2.0)
```

```python
# Criando uma grid
plt.plot(x,y,label='$Y(X)=(X-2)^2$')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('y(x)')
plt.grid(visible=True, linestyle = '--' , which = 'major' , color = 'grey') # O parametro B foi renomeado para visible.
plt.legend()
plt.show()

```

```python
# Alterando a cor de fundo
plt.plot(x,y,label='$Y(X)=(X-2)^2$')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('y(x)')
plt.grid(visible=True , linestyle = '--' , which = 'major' , color = 'red')
plt.legend()
plt.show()
```
