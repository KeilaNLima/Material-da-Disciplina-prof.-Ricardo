```python
# Importando as Bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando variavel x
X = np.linspace(-5,5,101)
```

```python
# Criando variavel y
Y = (X-2.0)*(X+2.0)
```

```python
# Criando gráfico
plt.plot(X,Y)
plt.xlabel('x')
plt.ylabel('y')
plt.fill_between(X,Y,)
plt.show()
```
