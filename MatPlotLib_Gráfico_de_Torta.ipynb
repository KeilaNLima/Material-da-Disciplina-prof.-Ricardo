```python
# Importando bibliotecas

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando labels

labels = ('A1','A2','A3','A4','A4')
```

```python
# Criando listas com valores
valores = [10,45,32,56,23]
```

```python
plt.pie(valores,labels=labels,startangle=90)
plt.show()
```

```python
plt.pie(valores,labels=labels,startangle=90,autopct='%1.1f%%')
plt.show()
```

```python
# Inserindo porcentagem
plt.pie(valores,labels=labels,startangle=90, autopct='%1.1f%%' )
plt.title('Frequ√™ncia dos dados do curso')
plt.show()
```

```python
plt.pie(valores,labels=labels,startangle=180, autopct='%1.1f%%',shadow=True)
plt.show()
```

```python
plt.pie(valores,labels=labels,startangle=90, autopct='%1.1f%%',shadow=True,explode=(0,0,1,0,0,))
plt.show()
```

```python
plt.pie(valores,labels=labels,startangle=90, autopct='%1.1f%%',shadow=True,explode=(0,0,1,2,0,))
plt.show()
```
