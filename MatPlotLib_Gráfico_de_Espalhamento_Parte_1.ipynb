```python
# Importando Bibliotecas

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando variavel x

x = np.linspace(0,5,11)
```

```python
# Criando funçoes y(x) z(x)

y = np.power(x-2,2.0)
```

```python
z = np.exp(x/2.0)
```

```python
# Criando grafico y com funçao x

plt.scatter(x,y)
plt.show()
```

```python
# Inserindo eixos no grafico

plt.scatter(x,y)
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.show()
```

```python
# Colocando titulo no grafico

plt.scatter(x,y)
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('y(x)')
plt.show()
```

```python
# Colocando dois graficos na mesma imagem

plt.scatter(x,y)
plt.scatter(x,z)
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('y(x), z(x)')
plt.show()
```

```python
# Alterando simbolo

plt.scatter(x,y, marker='.')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('y(x)')
plt.show()
```

```python
# Dois graficos com simbolo diferente

plt.scatter(x,y, marker='p')
plt.scatter(x,z,marker='v')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('y(x)')
plt.show()
```

```python
# Alterando tamanho dos pontos

plt.scatter(x,y,s=10)
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('y(x)')
plt.show()
```

```python
plt.scatter(x,y,s=20,label='y(x)')
plt.scatter(x,z,s=50,label='z(x)')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('y(x), z(x)')
plt.legend()
plt.show()
```

```python
# Inserindo Transparencia

plt.scatter(x,y,s=20,label='y(x)', alpha=0.5)
plt.scatter(x,z,s=50,label='z(x)',alpha=0.4)
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('y(x), z(x)')
plt.legend()
plt.show()
```
