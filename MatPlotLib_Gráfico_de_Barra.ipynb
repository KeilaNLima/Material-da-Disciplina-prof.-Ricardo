```python
# Importando Bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando labels
labels = ['A1', 'A2','A3','A4','A5' ]
```

```python
# Criando lista com valores
valores = [10,45,32,56,23]
```

```python
# Criando grafico de barras
plt.bar(labels,valores)
```

```python
# Inserindo nomes nos eixos
plt.bar(labels,valores)
plt.xlabel('ID')
plt.ylabel('valor')
```

```python
# Alterando cor do grafico
plt.bar(labels,valores,color= 'green')
plt.xlabel('ID')
plt.ylabel('valor')
```

```python
# Gr√°fico na horizontal
plt.barh(labels,valores,color= 'green')
plt.xlabel('ID')
plt.ylabel('valor')
```

```python
# inserindo valores nas barras
plt.bar(labels,valores,color='green')
plt.xlabel('ID')
plt.ylabel('valor')

for x,y in zip(labels,valores):
  plt.text(x, y+0.05, '%d' % y, ha='center', va= 'bottom' )
```

```python
# Inserir 2 gr√°fico de barra
valores2 = [8,30,60,21,30]
```

```python
largura = 0.3
pos1 = np.arange(len(valores))
pos2 = [x + largura for x in pos1]

plt.bar(pos1, valores, color= 'green', width=largura, label='barra 1')
plt.bar(pos2, valores2, color= 'black', width=largura, label='barra 2')
plt.legend()
```

```python
plt.bar(pos1,valores,label = "Barra 1")
plt.bar(pos1,valores2,label = "Barra 2",bottom = valores)
```

```python
plt.barh(pos1,valores,label = "Barra 1")
plt.barh(pos1,valores2,label = "Barra 2",left = valores)
```

```python
plt.barh(pos1,valores,label = "Barra 1")
plt.barh(pos1,valores2,label = "Barra 2",left = valores, linewidth = 2, hatch = '+' )
```
